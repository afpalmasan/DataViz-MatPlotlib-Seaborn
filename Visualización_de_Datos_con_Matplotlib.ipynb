## INTERFAZ ORIENTADA A OBJETOS

---

*texto en cursiva*

Método orientado a objetos en Matplotlib
Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes grá ficas(axes), de las cuales cada una posee sus propios ejes(axis).

La figura representa el todo, dentro de ella vamos a configurar gráficas las cuales contienen diferentes ejes.

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r:')
axes2.set_facecolor('gray')

fig.show()
```

## SUBPLOTS

---



Con subplots se puede trabajar en un arreglo de gráficas a las cuales se accede a través de los índices.

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r:')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2) ## se puede definir lso axes con una tupla
ax1.plot(x,y,'b')
ax2.plot(y,x,'r:')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=4)
axes[0,0].plot(x, np.cos(x),'b')
axes[0,1].plot(x, np.sin(x),'y')
axes[0,2].plot(x, np.tan(x),'purple')
axes[0,3].plot(x, np.cos(x)**2,'black')

axes[1,0].plot(x, np.sin(y),'g:',)
axes[1,1].plot(x, np.sin(x)**2,'y--')
axes[1,2].plot(x, np.tan(y),'o')
axes[1,3].plot(x, np.cos(y)**2,'rs-')

fig.tight_layout()
```

## TITULOS, LABELS, SIZE

---



Para dar contexto a nuestros gráficos necesitamos usar títulos, leyendas, tamaño o etiquetas, para que nuestra gráfica tenga un contexto más amplio

metodo orientado a objetos





```python
fig, axes = plt.subplots(1,2,figsize=(5,5))
axes[0].plot(x,y, label='$Sin(x)$')
axes[0].set_title('Relacion x - y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('y')
axes[0].legend()

axes[1].plot(y,x, 'g', label='$Sin(y)$')
axes[1].set_title('Relacion y - x')
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
axes[1].legend()

fig.tight_layout()
```
