## INTERFAZ ORIENTADA A OBJETOS

---

*texto en cursiva*

Mtodo orientado a objetos en Matplotlib
Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes gr ficas(axes), de las cuales cada una posee sus propios ejes(axis).

La figura representa el todo, dentro de ella vamos a configurar grficas las cuales contienen diferentes ejes.

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r:')
axes2.set_facecolor('gray')

fig.show()
```

## SUBPLOTS

---



Con subplots se puede trabajar en un arreglo de grficas a las cuales se accede a travs de los ndices.

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r:')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2) ## se puede definir lso axes con una tupla
ax1.plot(x,y,'b')
ax2.plot(y,x,'r:')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=4)
axes[0,0].plot(x, np.cos(x),'b')
axes[0,1].plot(x, np.sin(x),'y')
axes[0,2].plot(x, np.tan(x),'purple')
axes[0,3].plot(x, np.cos(x)**2,'black')

axes[1,0].plot(x, np.sin(y),'g:',)
axes[1,1].plot(x, np.sin(x)**2,'y--')
axes[1,2].plot(x, np.tan(y),'o')
axes[1,3].plot(x, np.cos(y)**2,'rs-')

fig.tight_layout()
```

## TITULOS, LABELS, SIZE

---



Para dar contexto a nuestros grficos necesitamos usar ttulos, leyendas, tamao o etiquetas, para que nuestra grfica tenga un contexto ms amplio

metodo orientado a objetos





```python
fig, axes = plt.subplots(1,2,figsize=(5,5))
axes[0].plot(x,y, label='$Sin(x)$')
axes[0].set_title('Relacion x - y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('y')
axes[0].legend()

axes[1].plot(y,x, 'g', label='$Sin(y)$')
axes[1].set_title('Relacion y - x')
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
axes[1].legend()

fig.tight_layout()
```

sin usar orientacion a objetos

```python
plt.plot(x,y, label='sin(x)')
plt.title("SENO DE X")
plt.xlabel('X')
plt.ylabel('y')
plt.legend(loc='lower left')
```

```python
print(plt.style.available) ## estilos disponibles en plt
```

```python
for i in plt.style.available: ## ver etilos disonibles
  plt.style.use(i)
  fig, ax = plt.subplots(figsize=(5,5))
  ax.set_title(i)
  ax.plot(x, y)
```

```python
fig, ax = plt.subplots(figsize = (6,6)) # estilo matlab
ax.plot(x,x+1, 'rs--')
ax.plot(x,x+2, 'bo--')
ax.plot(x,x+3, 'go:')
ax.plot(x,x+4, 'y.-')
```

Stylo Pyplot

```python
fig, ax = plt.subplots(figsize = (6,6)) # estilo matlab
ax.plot(x,x+1, color='green', alpha=0.8, linewidth=5, linestyle='-', marker='o')
ax.plot(x,x+2, color='blue',linewidth=8, linestyle='--', marker='v')
ax.plot(x,x+3, color='#22EEAA', linewidth=6,linestyle='dashed', marker='p')
ax.plot(x,x+4, color='#CCBB44', linewidth=2,linestyle=':', marker='p')
```

## BAR PLOT

---



El grfico de barras nos permite graficar variables categricas, es decir, variables de texto, lo que es muy importante en el mundo de la ciencia de datos y Matplotlib ofrece ciertas caractersticas que nos facilita la vida en cuanto a graficar este tipo de variables.

```python
country= ['INDIA','JAPAN', 'MEXIXO', 'COLOMBIA', 'GERMANY']
population = [1000, 800,900,1500,300]
plt.bar(country,population,width=0.5, color=['red','blue'])
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexixo','Colombia', 'Alemania'), rotation=45)
plt.title('POBLACION X PAIS')
plt.show()

```

```python
plt.barh(country,population)
plt.show()
```
