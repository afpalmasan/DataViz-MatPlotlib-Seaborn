## PyPlot Basico

---



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y,'rx-') # usamos plot para crrar un grafico, se pueden modificar los marcadores
plt.show()
```

```python
plt.plot(x,y,'bo-') # usamos plot para crrar un grafico, se pueden modificar los marcadores
plt.show()
```

```python
plt.plot(x,y,'rs-') # usamos plot para crrar un grafico, se pueden modificar los marcadores
plt.show()
```

```python
plt.plot(x,y,'yD:') # usamos plot para crrar un grafico, se pueden modificar los marcadores
plt.show()
```

```python
plt.hist(x,y,) #histografa
plt.show()
```

```python
plt.pie(x) #torta
plt.show()
```

```python
plt.boxplot(x) #torta
plt.show()
```

## SUBPLOT

---



```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(x,y)
plt.show()
```

```python
plt.subplot(1,2,1) # Grafico 1
plt.plot(x,y, 'r--')

plt.subplot(1,2,2)  # Gráfico 2
plt.pie(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--') # Plot 1
plt.plot(y,x, 'b:') #Plot 2

plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1) #Grafico 1
plt.plot(x,y,'r--') #Plot 1
plt.plot(y,x,'b:')  #Plot 2

plt.subplot(2,1,2) #Grafico 2
plt.pie(y)         #Plot 1
plt.show()
```

```python
plt.subplot(2,2,1) #Grafico 1
plt.plot(x,y,'r--') #Plot 1
plt.plot(y,x,'b:')  #Plot 2

plt.subplot(2,2,2) #Grafico 2
plt.pie(y)         #Plot 1
plt.show()

plt.subplot(2,2,3) #Grafico 3
plt.hist(y,x)         #Plot 1
plt.show()

plt.subplot(2,2,4) #Grafico 4
plt.triplot(y,x)         #Plot 1
plt.show()
```

Método orientado a objetos en Matplotlib
Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes grá ficas(axes), de las cuales cada una posee sus propios ejes(axis).

La figura representa el todo, dentro de ella vamos a configurar gráficas las cuales contienen diferentes ejes.
