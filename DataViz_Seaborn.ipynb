## SEABORN

---





SET : Set es un método de Seaborn que permite configurar el estilo, fuente y color de las gráficas.

```python
import seaborn as sns # Importar libreria de forma estandar como sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','b','C'], y=[1,3,2])
plt.show()
```

## Configurar fuentes y estilos
Configura el estilo, la paleta de colores, la fuente y el tamaño de la fuente de la gráfica.

```python
sns.set(style='dark',)
sns.barplot(x=['A','b','C'], y=[1,3,2])
plt.show()
```

## PARAMETRO USADO EN SEABORN

---



Seaborn tiene una gran variedad de gráficos, pero también tiene ciertos parámetros para cada gráfico, vamos a ver cuáles son los más comunes:

```python
tips = sns.load_dataset('tips') ## dataset de seaborn
tips
```

```python
sns.displot(data=tips, x='total_bill'); #Puedes usar ’ ; ’ al final de la linea de código para remplazar el 'plt.show()'
```

Agregamos una variable ‘Y’, pero seaborn reconoce que en gráfico de frecuencias no es la mejor opción y lo adapta automáticamente



```python
sns.displot(data=tips, x='total_bill', y='tip');
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex'); # Hacer agrupamiento por una variable hue
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde'); # Cambiar el tipo de gráfica kind
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=True, palette='Spectral', alpha= .5);#Quitar la legenda, cambiar la paleta de colores y su transparencia
```

## GRAFICOS DE DSITRIBUCION

---



Histograma

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='probability', multiple='dodge');# cumulative ver comportamiento acumulativo de frecuencias
```

Grafico de Densidad

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False);
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True);
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True, bw_adjust=1);
```

Grafico scd Plot

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
```

```python
sns.displot(data=tips, x='tip', hue='sex',multiple='stack')
```

```python
tips = sns.load_dataset("tips") # convertimos el dataset de tips a dataframe
tips['weekend'] =  (tips['day'] == 'Sun') | (tips['day'] == 'Sat') # Agregar columna weekend
sns.displot(data=tips,x='tip',hue='weekend', kde=True, multiple='dodge')
```

```python
sns.displot(data=tips,x='total_bill',hue='weekend',  kde=True, multiple='dodge')
```

```python
#distribucion de total de la cuenta vs propinas agrupado por fn de semana
sns.displot(data=tips,x='total_bill', y="tip", hue="weekend", kind="kde")
```

## GRAFICOS DE RELACION

---



Seaborn maneja los gráficos de relación entre distintas variables numéricas para hacer visualizaciones. Veamos los distintos gráficos que tiene Seaborn para trabajar este tipo de relaciones.

```python
markers = {"Lunch": 'D', "Dinner": 's'}
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size');
plt.legend(loc= 'center', bbox_to_anchor= (1.12, 0.5)) # ajustar tamaño de legenda
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', col='time') # grafico de relacion por defaul ajusta el trafico mas conveniente
```

# JOINPLOT / PAIRPLOT

---



Jointplot y Pairplot son funcionalidades de Seaborn, porque con un solo comando se puede generar varios tipos de gráficos.

JoinPlot

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip');
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex');
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist');
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True); #muestra una pequeña tabla para la gráfica exterior
```

PairPlot

```python
sns.pairplot(data=tips); # grafica relacion entre variable numericas
```

```python
sns.pairplot(data=tips, hue='sex');#segmentacion por una variable categorica
```

```python
sns.pairplot(data=tips, hue='sex', corner=True); #Solo muestra una esquina del pairplot, eliminando así, las gráficas repetidas
```
